# Our porfolio excess measurement

```python
import os
import pandas as pd


data = pd.DataFrame(pd.read_csv('https://assets.datacamp.com/production/repositories/1546/datasets/3d9b734fea954b629d2477ef48c36525dfecf6e0/FamaFrenchFactors.csv', parse_dates = ['Date']))\
.set_index('Date')
data
```

```python
# Calculate excess portfolio returns
data['Portfolio_Excess'] = data['Portfolio'] - data['RF']
data
```

```python

import matplotlib.pyplot as plt
# Plot returns vs excess returns
dataReturns = ((1+data[['Portfolio','Portfolio_Excess']]).cumprod()-1)
dataReturns.plot()
plt.show()
```
